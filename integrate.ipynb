calculus in python

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

arr = np.linspace(0,10)

x = sym.Symbol('x')
y = np.sin(arr)
dydx = np.gradient(y)*5

y_intsym = sym.integrate(sym.sin(x) , x)
dydxsym = sym.diff(sym.sin(x),x)

plt.plot(arr,y)
plt.plot(arr,dydx)
print("integration of sin(x) = ",y_intsym,"+c")
print("diff of sin(x) = ",dydxsym)
```
